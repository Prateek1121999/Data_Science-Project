```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.pipeline import Pipeline

# Define the features (X) and the target variable (y) for training data
X_train = df_train[['Square Footage', 'Number of Occupants', 'Appliances Used', 'Average Temperature']]
y_train = df_train['Energy Consumption']

# Define the features (X) and the target variable (y) for testing data
X_test = df_test[['Square Footage', 'Number of Occupants', 'Appliances Used', 'Average Temperature']]
y_test = df_test['Energy Consumption']

# Create a pipeline with feature scaling and feature selection
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('selector', SelectKBest(f_regression, k=4)),
    ('regressor', LinearRegression())
])

# Train the pipeline on the training data
pipeline.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = pipeline.predict(X_test)
```

```python
# Evaluate the model using mean squared error and R-squared
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f'Mean Squared Error: {mse:.2f}')
print(f'R-squared: {r2:.2f}')

# Use the model to make predictions on new data
new_data = pd.DataFrame({
    'Square Footage': [1000, 2000, 3000],
    'Number of Occupants': [5, 10, 15],
    'Appliances Used': [10, 20, 30],
    'Average Temperature': [20, 25, 30]
})
new_pred = pipeline.predict(new_data)
print(f'Predicted Energy Consumption: {new_pred}')
```

```python
from sklearn.metrics import accuracy_score
import numpy as np
# Predict the values
y_pred = pipeline.predict(X_test)

# Since this is a regression problem, we need to convert the predictions to classes
# Let's assume we want to predict if the energy consumption is above or below the median
median = np.median(y_test)
y_pred_class = np.where(y_pred > median, 1, 0)
y_test_class = np.where(y_test > median, 1, 0)

# Calculate the accuracy score
accuracy = accuracy_score(y_test_class, y_pred_class)

# Print the accuracy score
print(f'Accuracy Score: {accuracy:.2f}')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import pandas as pd

# Assuming df_train and df_test are your training and testing data DataFrames
# Define features (X) and target variable (y) for training and test data
X_train = df_train[['Square Footage', 'Number of Occupants', 'Appliances Used', 'Average Temperature']]
y_train = df_train['Energy Consumption']
X_test = df_test[['Square Footage', 'Number of Occupants', 'Appliances Used', 'Average Temperature']]
y_test = df_test['Energy Consumption']

# Create a pipeline with scaling and a linear regression model
pipeline = Pipeline([
    ('scaler', StandardScaler()),  # Standardize the data (important for regression)
    ('regressor', LinearRegression())  # Linear Regression Model
])

# Train the model on the training data
pipeline.fit(X_train, y_train)

# Predict on the testing data
y_pred = pipeline.predict(X_test)

# Evaluate the model performance
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print model evaluation metrics
print(f'Mean Squared Error: {mse:.2f}')
print(f'R-squared: {r2:.2f}')

# Use the model to predict new data
new_data = pd.DataFrame({
    'Square Footage': [1000, 2000, 3000],
    'Number of Occupants': [5, 10, 15],
    'Appliances Used': [10, 20, 30],
    'Average Temperature': [20, 25, 30]
})

new_pred = pipeline.predict(new_data)
print(f'Predicted Energy Consumption: {new_pred}')
```

```python
from sklearn.ensemble import RandomForestRegressor

# Create a pipeline with scaling and RandomForestRegressor
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('regressor', RandomForestRegressor(n_estimators=100, random_state=42))  # Using 100 trees
])

# Train the model on the training data
pipeline.fit(X_train, y_train)

# Predict on the testing data
y_pred = pipeline.predict(X_test)

# Evaluate the model performance
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print model evaluation metrics
print(f'Mean Squared Error: {mse:.2f}')
print(f'R-squared: {r2:.2f}')
```
